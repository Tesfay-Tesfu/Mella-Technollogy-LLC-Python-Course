<h3 style="color:green;">Advantages of Jupyter Notebook (Why people choose it)</h3>

<ul>
  <li><span style="color:orange;"><strong>Interactive Output</strong></span><br>
  &emsp;Run code cells one by one and instantly see output (text, plots, images) inline.</li>

  <li><span style="color:teal;"><strong>Ideal for Data Science & Research</strong></span><br>
  &emsp;Perfect for step-by-step data analysis, machine learning, and scientific computing.</li>

  <li><span style="color:purple;"><strong>Combines Code + Documentation</strong></span><br>
  &emsp;Write explanations using <code>LaTeX</code> and Markdown alongside your code.</li>

  <li><span style="color:green;"><strong>Visualizations Built-in</strong></span><br>
  &emsp;Integrates smoothly with <code>matplotlib</code>, <code>seaborn</code>, <code>plotly</code>, etc., to display charts inline.</li>

  <li><span style="color:brown;"><strong>Cell-Based Execution</strong></span><br>
  &emsp;Run small code blocks independently for debugging or experimentation.</li>

  <li><span style="color:blue;"><strong>Great for Prototyping</strong></span><br>
  &emsp;Try ideas quickly without restarting the whole program.</li>

  <li><span style="color:magenta;"><strong>Web-Based Interface</strong></span><br>
  &emsp;Runs in your browser; no heavy IDE required.</li>


  <li><span style="color:red;"><strong>Widely Used in Education</strong></span><br>
