### Calculate the difference between 'hpF2' and 'h'F' columns represent as "Delta_F" column

```python
Delta_F = [hp - hf for hp, hf in zip(hpF2, hF)]
# Output file path for the new file
output_file_path1 = 'Output_file1.txt'
# Write data to the new file with the additional column
with open(output_file_path1, 'w') as output_file1:
    # Write header line
    output_file1.write("UT\tLT\th'F\tfoF2\thpF2\tSF\tES\tdifference\n")
    # Write data lines with the calculated difference
    for i in range(len(UT)):
        output_file1.write(f"{UT[i]}\t{LT[i]}\t{hF[i]}\t{foF2[i]}\t{hpF2[i]}\t{SF[i]}\t{ES[i]}\t{Delta_F[i]}\n")
print(f" Hi Tesfay, your New_Data may be Saved Please Check !!! {output_file_path1}")
```

```python
def Tesfay_hours_to_hhmmss(decimal_hours):
    hhmmss_list = []
    for hours in decimal_hours:
        # Convert decimal hours to hours, minutes, and seconds
        total_seconds = int(hours * 3600)
        hours_part = total_seconds // 3600
        minutes_part = (total_seconds % 3600) // 60
        seconds_part = total_seconds % 60

        # Format as HH:MM:SS and append to the result list
        formatted_time = f"{hours_part:02d}:{minutes_part:02d}:{seconds_part:02d}"
        hhmmss_list.append(formatted_time)

    return hhmmss_list

# Example usage:
# decimal_hours = [4.5, 8.25, 2.75]
hhmmss_times = Tesfay_hours_to_hhmmss(LT)
print(hhmmss_times)  # Output: ['04:30:00', '08:15:00', '02:45:00']
```
