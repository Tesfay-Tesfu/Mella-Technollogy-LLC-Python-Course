```python
num_inputs = 4
Total_price = []
Profits = []
Initial_Price = []
for i in range(num_inputs):
    P = int(input("Please enter your initial price here: "))
    Initial_Price.append(P)
    P_factory = P + (P * 45) / 100
    Distributor_profit = (P_factory) * 12 / 100
    Profits.append(Distributor_profit)
    P_at_distributor = P_factory + (P_factory * 12) / 100
    Total_price.append(P_at_distributor)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame(list(zip(Initial_Price,Total_price, Profits)),columns =['Initial_Price','Total_price', 'Profits'])
df
```

```python
plt.plot(df["Initial_Price"],df["Total_price"],'r-o', label = 'total')
plt.plot(df["Initial_Price"],df["Profits"],'k-*', label = 'price')
plt.legend()
#plt.plot(df["Initial_Price"],df["Total_price"],'g^')

plt.show()
```
